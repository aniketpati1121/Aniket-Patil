Bar Plot

```python
import matplotlib.pyplot as plt

x= ["python","c","c++","java"]
y= [96,89,54,68]
plt.xlabel("language",fontsize=25)
plt.ylabel("percent",fontsize=25)
plt.title("aniket patil",fontsize=25)
#=["y","b","r","g"]
plt.bar(x,y,color=c,width=0.4,align="edge",edgecolor="r",linewidth=10)

plt.show

#align function used to Align the ylabels of subplots with the same subplots
#row or column if label alignment is being done automatically.


#edgecolor="r" this function draw the bar line egde to color

#linewidth=10 draw the outer side bar
```

```python
import matplotlib.pyplot as plt

x= ["python","c","c++","java"]
y= [96,89,54,68]
z= [22,34,14,45]
plt.xlabel("language",fontsize=25)
plt.ylabel("percent",fontsize=25)
plt.title("aniket patil",fontsize=25)
#=["y","b","r","g"]
plt.bar(x,y,color="b",width=0.4,label="aniketpati")
plt.bar(x,z,color="r",width=0.4,label="aniketpati1")

plt.legend()

plt.show

#plt.legend() this function draw infomation graph in a bar
```

```python
import matplotlib.pyplot as  plt
import numpy as np

x= ["python","c","c++","java"]
y= [96,89,54,68]
z= [22,34,14,45]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=10)
plt.ylabel("percent",fontsize=10)
plt.title("aniket patil",fontsize=10)
width=0.5
plt.bar(p,y,color="b",label="aniketpati")
plt.bar(p1,z,color="r",label="aniketpati1")

plt.xticks(p+width/2,x,rotation=24)
plt.legend()
plt.show
```

```python
#Horizontal bar graph
import matplotlib.pyplot as  plt
import numpy as np

x= ["python","c","c++","java"]
y= [96,89,54,68]
z= [22,34,14,45]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=10)
plt.ylabel("percent",fontsize=10)
plt.title("aniket patil",fontsize=10)
width=0.5
plt.barh(p,y,color="b",label="aniketpati")
plt.barh(p1,z,color="r",label="aniketpati1")

plt.xticks(p+width/2,x,rotation=24)
plt.legend()
plt.show
```

Matplotlib Scatter Plot

```python
import matplotlib.pyplot as  plt

day=[1,2,3,4,5,6,7,8]
no=[2,3,4,45,9,15,8,9]
colors=["r","b","g","y","b","y","r","b"]
sizes=[500,45,45,67,23,78,23,12]
plt.scatter(day,no,c=colors,s=sizes,marker="*")

plt.title("Scatter plot",fontsize=10)
plt.xlabel("Day",fontsize=10)
plt.ylabel("no",fontsize=10)



plt.show()
```

```python
import matplotlib.pyplot as  plt

day=[1,2,3,4,5,6,7,8]
no=[2,3,4,45,9,15,8,9]
colors=["r","b","g","y","b","y","r","b"]
sizes=[500,45,45,67,23,78,23,12]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.3,marker="*")

plt.title("Scatter plot",fontsize=10)
plt.xlabel("Day",fontsize=10)
plt.ylabel("no",fontsize=10)



plt.show()
```

```python
import matplotlib.pyplot as  plt

day=[1,2,3,4,5,6,7,8]
no=[2,3,4,45,9,15,8,9]
colors=[456,23,90,45,78,23,98,78]
sizes=[500,45,45,67,23,78,23,12]

plt.scatter(day,no,c=colors,s=sizes,cmap="Accent")
t=plt.colorbar()
t.set_label("ColorBar")

plt.title("Scatter plot",fontsize=10)
plt.xlabel("Day",fontsize=10)
plt.ylabel("no",fontsize=10)



plt.show()
```

* Pie Plot in Matplotlib




```python
import matplotlib.pyplot as  plt

x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.1,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",
       shadow=True,radius=0.7,labeldistance=1.1)
plt.title("Data Analysis")
plt.legend(loc=2)
plt.show()
```

* Stems Plots with Matplotlib

```python
import matplotlib.pyplot as   plt

x = [1,2,3,4,5,6]
y = [2,2,3,5,4,6]

plt.stem(x,y,linefmt=":",markerfmt="r",bottom =9,basefmt= "g",
         label="python")
plt.legend()
plt.show()
```

# Whisker Plot in Matplotlib

* Stack Plot

```python
import matplotlib.pyplot as   plt

x = [9,2,3,4,5]

area = [1,2,7,4,5]
area1 = [3,4,2,4,5]
area2 = [1,3,3,4,5]
area3 = [1,3,3,4,3]

plt.stackplot(x,area,area1,area2,area3)


plt.show()
```

```python
import matplotlib.pyplot as   plt

x = [9,2,3,4,5]

area = [1,2,7,4,5]
area1 = [3,4,2,4,5]
area2 = [1,3,3,4,5]
area3 = [1,3,3,4,3]

l =["area1","area2","area3"]

plt.stackplot(x,area,area1,area2,area3,labels=l,colors=["r","g","b","m"],
             baseline="sym")


plt.title("PYTHON")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=1)
plt.show()
```

Step plot in Matplotlib

```python
import matplotlib.pyplot as   plt

x = [1,2,3,4,5,6]
y = [1,2,3,1,5,15]

plt.step(x,y,color="r",marker="o",label="python")


plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("python")
plt.grid()
plt.show()
```

File Between Plot in Matplolib

```python
import matplotlib.pyplot as   plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]


plt.plot(x,y,color="r")

plt.fill_between(x,y)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
